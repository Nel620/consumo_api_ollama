```python
import time
from ollama import Client

client = Client(host='http://api.entornoandres.me')
datos_entrada = "Ventas Q1: $5000, Q2: $7000. Calcula el porcentaje de crecimiento."

# 1. Iniciar cronómetro
inicio = time.time()
print("⏳ Procesando petición...")

# 2. Ejecutar inferencia (Cambia 'llama3' por 'phi3' para comparar)
response = client.chat(model='llama3', messages=[
  {
    'role': 'user',
    'content': f'Actúa como analista de datos. Procesa esto: {datos_entrada}',
  },
])

# 3. Detener cronómetro y calcular
fin = time.time()
tiempo_total = fin - inicio

print(response['message']['content'])
print(f"\n⏱️ Tiempo exacto de ejecución: {tiempo_total:.2f} segundos")
```
